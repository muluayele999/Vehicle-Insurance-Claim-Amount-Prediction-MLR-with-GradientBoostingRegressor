# Shapiro for Normality test of All Numeric Columns<a id='shapiro-norm'></a>

```python
#Normality test/Distribution test of all columns/variables
from scipy.stats import shapiro
stat,pvalue = shapiro(Motor['vehage'])
if pvalue >= 0.05:
    print("Null Hypothesis (H0): vehage is Normally distributed")
else:
    print("Alternate Hypothesis (H1): vehage is not Normally Distributed")
print("Confidence Level : {} %".format(((1- pvalue)*100)))
print('Statistics=%.3f, p=%.3f' % (stat, pvalue))

#*D'Agostino's Skewness/K^2 test(normality)
from scipy.stats import normaltest
stat,pvalue = normaltest(Motor['vehage'])
if pvalue >= 0.05:
    print("Null Hypothesis (H0): vehage is Normally distributed")
else:
    print("Alternate Hypothesis (H1): vehage is not Normally Distributed")
print("Confidence Level : {} %".format(((1- pvalue)*100)))
print('Statistics=%.3f, p=%.3f' % (stat, pvalue))

#*Anderson Darling Test(normality)
from scipy.stats import anderson
Result=anderson(Motor['vehage'])
p=0
for i in range(len(Result.critical_values)):
    SL, CV = Result.significance_level[i], Result.critical_values[i]
    if Result.statistic < Result.critical_values[i]:
        print("Null Hypothesis (H0): '%.3f : %.3f, vehage is Normally distributed"%(SL,CV))
    else:
        print("Alternate Hypothesis (H1): '%.3f : %.3f, vehage is not Normally Distributed"%(SL,CV))
print("Confidence Level : {} %".format(((1- Result.statistic)*100)))
print('Statistic: %.3f' % Result.statistic)
```

# Spearman Correlation Test for Measures of Association between non-normal Datas<a id='spear-corr'></a>
1. As all columns of Motor DataFrame are not normally distributed then we will use Spearman Correlation
2. Regression only assumes normality for the outcome (Label) variable.
3. Non-normality in the predictors (Features) May create a nonlinear relationship between them and the Y (Label)
4. Try to transform the Y (Label) to Normal distribution
5. If you have lots of non-normal Features (Predictors) that can cause heterogeneity of Variance which is the bigger problem
7. What are the non-parametric alternatives of Multiple Linear Regression ?? It is non Linear regression
8. You either Try to transform the Y (Label) to Normal distribution or perform non Linear regression

```python
#Spearman Correlation test (When two Continuous Variables have Non linear Relationship)
#Correlation between Rank of "claimamt" & Rank of "vehage"
#Distribution of Ranking Squared Differences
#Scipy ---> Scientific Python ---> All formulaes related to math and stats to be applied
#Acceptable p-value for my hypothesis is 0.05
from scipy.stats import spearmanr
correlation,pvalue = spearmanr(Motor_Clean['claimamt'] , Motor_Clean['vehage'])
print('Spearmans correlation coefficient:%.3f' % correlation)
print(pvalue)
print("Confidence Level : {} %".format(((1- pvalue)*100)))
if pvalue <= 0.05:
    print("Alternate Hypothesis (H1) - claimamt and vehage have non linear relationship")
else:
    print("Null Hypothesis (H0)- claimamt and vehage have no non linear relationship")
    
#We will Perform Kendall's Rank Correlation (When two Continuous Variables have Non linear Relationship)
#To find Correlation between Rank of "claimamt" & Rank of "vehage" we will Perform Kendall's Rank Correlation
#Correlation between Rank of "claimamt" & Rank of "vehage"
#Kendall's Tau=(Concordant-Discordant/Concordant+Discordant)
from scipy.stats import kendalltau
correlation,pvalue = kendalltau(Motor_Clean['claimamt'] , Motor_Clean['vehage'])
print("Kendall's Rank correlation coefficient:%.3f" % correlation)
print(pvalue)
print("Confidence Level : {} %".format(((1- pvalue)*100)))
if pvalue <= 0.05:
    print("Alternate Hypothesis (H1) - claimamt and vehage have non linear relationship")
else:
    print("Null Hypothesis (H0)- claimamt and vehage have no non linear relationship")

#Detailed Test
from scipy.stats import linregress
linregress(Motor_Clean['claimamt'] , Motor_Clean['vehage'])
```

# Regression Evaluation Metrics<a id='mod-eval'></a>


Here are three common evaluation metrics for regression problems:

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) is the mean of the squared errors:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparing these metrics:

- **MAE** is the easiest to understand, because it's the average error.
- **MSE** is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
- **RMSE** is even more popular than MSE, because RMSE is interpretable in the "y" units.

All of these are **loss functions**, because we want to minimize them.
